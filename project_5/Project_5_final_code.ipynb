```python
# Import PVI data
df_PVI=pd.read_csv('Cook PVI by State - PVI 2.csv') 
df_PVI.set_index('State',inplace=True)
df_PVI=df_PVI.unstack()
df_PVI=pd.DataFrame(df_PVI)
df_PVI=df_PVI.reset_index()
df_PVI.rename(columns={'level_0':'Year',0:'PVI'},inplace=True)
df_PVI.sort_values(['Year','State'],ascending=[False,True],inplace=True)
df_PVI=df_PVI.reset_index(drop=True)
df_PVI
```

```python
# Import econ data
df_GDP=pd.read_excel('State GDP per capita growth.xls') 
df_GDP.columns=df_GDP.iloc[4]
df_GDP=df_GDP.iloc[6:57,1:]
df_GDP.rename(columns={'GeoName':'State','1999-2000':'2000','2007-2008':'2008','2015-2016':'2016'},inplace=True)
df_GDP.set_index('State',inplace=True)
df_GDP=df_GDP.unstack()
df_GDP=pd.DataFrame(df_GDP)
df_GDP=df_GDP.reset_index()
df_GDP.rename(columns={4:'Year',0:'Per Capita GDP Growth'},inplace=True)
df_GDP.sort_values(['Year','State'],ascending=[False,True],inplace=True)
df_GDP=df_GDP.reset_index(drop=True)
df_GDP
```

```python
# Import all other data (turnout, etc.)
df=pd.read_csv('Election Data - All data.csv') 
df.replace(to_replace={'Yes'}, value={1}, inplace=True)
df['Dem caucus'].fillna(0, inplace=True)
df['GOP caucus'].fillna(0, inplace=True)
df['Dem turnout'].replace(0,np.nan,inplace=True)
df['GOP turnout'].replace(0,np.nan,inplace=True)
df
```

```python
# Merge datasets
df=pd.concat([df,df_PVI.iloc[:,2]],axis='columns',join='outer',keys=None) 
df=pd.concat([df,df_GDP.iloc[:,2]],axis='columns',join='outer',keys=None)
df
```

```python
# Create dummy variable to reflect party of incumbent president
df['Dem incumbent']=np.nan 
for row in range(len(df)):
    if df.iloc[row,0]==2016 or df.iloc[row,0]==2000:
        df['Dem incumbent'][row]=1
    else:
        df['Dem incumbent'][row]=0
        
# Compute adjusted turnout variables        
df['Dem adj turnout']=df['Dem turnout']/(1+2*df['PVI'])
df['GOP adj turnout']=df['GOP turnout']/(1-2*df['PVI'])
df.drop(columns={'Dem turnout','GOP turnout'},inplace=True)

# Compute variables to reflect caucus/primary status
df['Dem caucus turnout']=df['Dem caucus']*df['Dem adj turnout']
df['GOP caucus turnout']=df['GOP caucus']*df['GOP adj turnout']

# Compute variable to reflect status of nomination contest
df['Dem fin turnout']=df['Dem finished']*df['Dem adj turnout']
df['GOP fin turnout']=df['GOP finished']*df['GOP adj turnout']

# Rescale GDP variable and create an interaction with incumbent president party
df['Per Capita GDP Growth']=df['Per Capita GDP Growth']/100
df['Inc Per Capita GDP Growth']=df['Per Capita GDP Growth']*df['Dem incumbent']

# Compute national polling variable
df['National Polling']=np.nan
for row in range(len(df)):
    if df.iloc[row,0]==2016:
        df['National Polling'][row]=0.032
    elif df.iloc[row,0]==2008:
        df['National Polling'][row]=0.076
    else:
        df['National Polling'][row]=-0.021
        
# Compute binary status of state-level electoral outcome        
df['Dem vs. Rep']=np.nan
for row in range(len(df)):
    if df.iloc[row,2]>0:
        df['Dem vs. Rep'][row]=1
    else:
        df['Dem vs. Rep'][row]=0
```

```python
natcorr=np.linspace(0,1,21)
for k in natcorr:
    success=0
    for t in range(500000):
        national_error=random.random()
        outcomes=[]
        for e in state_probs:
            if (1-k)*random.random()+k*national_error < e:
                outcomes.append(1)
            else:
                outcomes.append(0)
        if outcomes==state_outcomes:
            success+=1
    print (round(k,2), round(success/100,6))
```

```python
# Use a national correlation factor of 0.12
```

```python
# Create function to correlate the state outcomes when running simulations.
def natl_corr(prob,national_error,national_share):
    if (1-national_share)*random.random()+national_share*national_error < prob:
        return 1
    else:
        return 0
```

```python
# Set up a Monte Carlo process to compute probability of each candidate winning electoral college.
```

```python
reg=2000 # Actually, C=2000 is probably closer to ideal.
natl=0.12
n_sim=100000 # Number of simulations to run for each model.
```

```python
# Each X represents a different model. Only include one; comment out the others.
y=df['Dem vs. Rep']
X=df[['PVI','Per Capita GDP Growth','Inc Per Capita GDP Growth']]
#X=df[['PVI','Per Capita GDP Growth','Inc Per Capita GDP Growth','Dem incumbent']]
#X=df.drop(columns=['Year','State','Dem-Rep %','Dem caucus','GOP caucus','Dem finished','GOP finished',
#                   'Electoral Votes','National Polling','Dem vs. Rep','Dem incumbent'])
#X=df.drop(columns=['Year','State','Dem-Rep %','Dem caucus','GOP caucus','Dem finished','GOP finished',
#                   'Electoral Votes','National Polling','Dem vs. Rep'])
ssX=StandardScaler()
X=ssX.fit_transform(X)
X_train=X[49:] # Indices are set manually to separate exactly into 2016 year-states vs. 2000 and 2008.
X_test=X[:49]
y_train=y.iloc[49:]
y_test=y.iloc[:49]
```

```python
lr = LogisticRegression(C=reg)
lr.fit(X_train, y_train) # Fit logistic regression model to 2000 and 2008 data
y_prob = lr.predict_proba(X_test) # Get state probabilities for 2016 election
y_prob=[e[-1].round(3) for e in y_prob] # Put probabilities into a list
pred_results = list(zip(df.iloc[:49,1],df.iloc[:49,7],y_prob,y_test.round(3))) # Zip state names, electoral votes, probabilities, outcomes
model_check=[]
for e in pred_results:
    model_check.append(list(e)) # Put the zip in a new list
model_check.append(['North Dakota',3,0.0,0.0]) # Add values for ND and WY since missing for 2016
model_check.append(['Wyoming',3,0.0,0.0])
for e in model_check: # Add true/false for model predictions vs. actual
    if abs(e[2]-e[3])<0.5:
        e.append(True)
    else:
        e.append(False)
model_check.sort(key=lambda state: state[2]) # Sort by state probabilities
model_check.sort(key=lambda state: state[-1]) # Then sort by true/false
print(brier_score_loss(y_test,y_prob)) # Return the Brier score for the model overall
for e in model_check: # Print each state in the list
    print(e)
```
